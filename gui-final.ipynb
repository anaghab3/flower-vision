```python
def open_img():
    # Select the Imagename  from a folder
    x = openfilename()
    print(type(x))
 
    # opens the image
    img = Image.open(x)
    matches = best_matches(x)
    names = helper(x)
    print(names)
     
    # resize the image and apply a high-quality down sampling filter
    img = img.resize((250, 250), Image.ANTIALIAS)
 
    # PhotoImage class is used to add image to widgets, icons etc
    img = ImageTk.PhotoImage(img)
  
    # create a label
    panel = Label(window, image = img)
     
    #set the image as img
    panel.image = img
    panel.grid(row = 3, column = 0)
    
    label12 = Label(window, text = "Our matches:", bg="white", fg="black", font=("Baloo Chettan 2", 30)).grid(row = 4, column = 0, sticky=W)
    
    label2 = Label(window, text = names[0], bg="white", fg="black", font="Playbill").grid(row = 5, column = 0, sticky=W)
    match1 = Image.open(matches[0])
    match1 = match1.resize((200, 200), Image.ANTIALIAS)
    match1 = ImageTk.PhotoImage(match1)
    panel2 = Label(window, image = match1)
    panel2.image = match1
    panel2.grid(row = 6, column = 0)
    
    label3 = Label(window, text = names[1], bg="white", fg="black", font="Playbill").grid(row = 5, column = 1, sticky=W)
    match2 = Image.open(matches[1])
    match2 = match2.resize((200, 200), Image.ANTIALIAS)
    match2 = ImageTk.PhotoImage(match2)
    panel3 = Label(window, image = match2)
    panel3.image = match2
    panel3.grid(row = 6, column = 1)
    
    label4 = Label(window, text = names[2], bg="white", fg="black", font="Playbill").grid(row = 5, column = 2, sticky=W)
    match3 = Image.open(matches[2])
    match3 = match3.resize((200, 200), Image.ANTIALIAS)
    match3 = ImageTk.PhotoImage(match3)
    panel4 = Label(window, image = match3)
    panel4.image = match3
    panel4.grid(row = 6, column = 2)
    
    label5 = Label(window, text = names[3], bg="white", fg="black", font="Playbill").grid(row = 13, column = 2, sticky=W)
    match4 = Image.open(matches[3])
    match4 = match4.resize((200, 200), Image.ANTIALIAS)
    match4 = ImageTk.PhotoImage(match4)
    panel5 = Label(window, image = match4)
    panel5.image = match4
    panel5.grid(row = 14, column = 2)
    
    label6 = Label(window, text = names[4], bg="white", fg="black", font="Playbill").grid(row = 13, column = 0, sticky=W)
    match5 = Image.open(matches[4])
    match5 = match5.resize((200, 200), Image.ANTIALIAS)
    match5 = ImageTk.PhotoImage(match5)
    panel6 = Label(window, image = match5)
    panel6.image = match5
    panel6.grid(row = 14, column = 0)
    
    label7 = Label(window, text = names[5], bg="white", fg="black", font="Playbill").grid(row = 13, column = 1, sticky=W)
    match6 = Image.open(matches[5])
    match6 = match6.resize((200, 200), Image.ANTIALIAS)
    match6 = ImageTk.PhotoImage(match6)
    panel7 = Label(window, image = match6)
    panel7.image = match6
    panel7.grid(row = 14, column = 1)
```

```python
def openfilename():
 
    # open file dialog box to select image
    # The dialogue box has a title "Open"
    filename = filedialog.askopenfilename(title ='"pen')
    return filename
```
